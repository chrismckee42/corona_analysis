```python
stats = []
```

```python
for i,item in enumerate(file_list):
    stats.append(pd.read_csv(item))
```

```python
# Creates a list of states, cleans it up and converts it to a dictionary
states = stats[0].Province_State[0:55]
states = states.to_list()
states.remove("Guam")
states.remove("Diamond Princess")
states.remove("Grand Princess")
states.remove("Puerto Rico")
states_dict = dict(zip(states,[*range(0,51)]))
states_dict
```

```python
test = [0] * 51
for i in range(len(stats[0])):
    if stats[0].Province_State[i] in states_dict:
        test[states_dict[stats[0].Province_State[i]]] = stats[0].Confirmed[i]
test
# test = np.array([stats[0].Confirmed,stats[2].Confirmed])
```

```python
dates = []
for file in file_list:
    dates.append(file[:-4])
dates
```

```python
confirmed = np.array([[0]*len(dates)]*len(states))
for i,x in enumerate(stats):
    for j in range(len(stats[i])):
        if stats[i].Province_State[j] in states_dict:
            confirmed[states_dict[stats[i].Province_State[j]]][i] = stats[i].Confirmed[j]
    
confirmed
```

```python
deaths = np.array([[0]*len(dates)]*len(states))
for i,x in enumerate(stats):
    for j in range(len(stats[i])):
        if stats[i].Province_State[j] in states_dict:
            deaths[states_dict[stats[i].Province_State[j]]][i] = stats[i].Deaths[j]
    
deaths
```

```python
confirmed_US = confirmed.sum(axis=0)
deaths_US = deaths.sum(axis=0)
```
